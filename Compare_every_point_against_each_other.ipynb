```python
import math
import numpy as np
import os
from ntpath import join
import pandas as pd
from scipy.io import loadmat
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nameh='ZM001_b'
namep='SM002_bal'
points=[1072,1076,1077,1098,1099,1116,1117,1118,1122,1123,1124,1140,1141,1142,1143,]
```

```python
#function for calcualating distance between two points
def distance(b1,b2):
    diss=math.sqrt(math.pow(b1[0]-b2[0],2)+math.pow(b1[1]-b2[1],2)+math.pow(b1[2]-b2[2],2))
    return diss
```

```python
#function to calculate norm of the reference plain
def norm(a,b,c):
    ab=np.array([(b[0]-a[0]),(b[1]-a[1]),(b[2]-a[2])])
    bc=np.array([(b[0]-c[0]),(b[1]-c[1]),(b[2]-c[2])])
    n=np.cross(ab,bc)
    return n
```

```python
#function for calculating koeficients of plain equations
def surface(a,b,c):
    n=norm(a,b,c)
    d=-(n[0]*b[0]+n[1]*b[1]+n[2]*b[2])
    f=[n[0],n[1],n[2],d]
    return f
```

```python
#function for calculating anble between reference plain and line conecting two points
def angle(b1,b2,n):
    b12=np.array([b2-b1])

    #calculating lenght of vectors
    nor_b12=np.linalg.norm(b12)
    nor_n=np.linalg.norm(n)

    kosphi=np.dot(b12,n)/(nor_b12*nor_n)
    ang_rad=math.acos(kosphi)
    ang=180*ang_rad/math.pi-90
    return ang
```

```python
#get the lenght of record
framesh=len(os.listdir(join(r'C:\Users\kuba8\MATLAB Drive\projekt oblicej\datapropouziti',nameh,nameh+'_coordinates')))
    
#prepare matrix for saving data
dfh=pd.DataFrame([])

for i in range(1,framesh):
    df=pd.DataFrame([])
    
    #load one photo
    datah=loadmat(join(r'C:\Users\kuba8\MATLAB Drive\projekt oblicej\datapropouziti',nameh,nameh+'_coordinates',nameh+'_coordinates_'+str(i)))
    datah=datah['model']
    
    #load time, when the photo was taken
    timeStemph=loadmat(join(r'C:\Users\kuba8\MATLAB Drive\projekt oblicej\datapropouziti',nameh,nameh+'_times',nameh+'_time_'+str(i)))
    
    for k in range(0,len(points)-1):
        for j in range(k+1,len(points)):

            name='points_'+str(points[k])+'_'+str(points[j])
            diss=distance(b1=datah[:,points[k]-1],b2=datah[:,points[j]-1])

            dfr = pd.DataFrame({name:diss},index=['frame_'+str(i)])
            df=pd.concat([df,dfr], axis=1)
        #include time to the last column
    time=pd.DataFrame({'time':timeStemph['timeStamp'][0][0]},index=['frame_'+str(i)])
    
    df=pd.concat([df,time],axis=1)
    dfh=pd.concat([dfh,df])
```
